```python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

print("--- Pokretanje Kvantnog Neurona (QNN Sloj) ---")

# 1. Pravimo kvantno kolo za naš neuron
kolo_qnn = QuantumCircuit(1)

# 2. Definišemo "Težinu neurona" (Weight) kao parametar ugla na Blohovoj sferi
# Ovo je broj koji će AI tokom učenja menjati da bi prepoznao hakera
tezina_neurona = Parameter('w')

# 3. Naš kvantni neuron radi tako što rotira kubit oko Y-ose za vrednost parametra 'w'
kolo_qnn.ry(tezina_neurona, 0)

# 4. SIMULIRAMO UČENJE: Zamisli da je AI nakon učenja shvatio da je idealan ugao 120 stepeni
# Prebacujemo 120 stepeni u radijane (to je oko 2.09 radijana)
izracunata_tezina = 120 * (np.pi / 180)

# Ubacujemo naučenu vrednost direktno u naš kvantni neuron
nauceno_kolo = kolo_qnn.assign_parameters({tezina_neurona: izracunata_tezina})

# 5. Izvlačimo konačno kvantno stanje nakon "učenja"
stanje_nakon_ucenja = Statevector.from_instruction(nauceno_kolo)

# 6. Vizuelizujemo gde je naš kvantni neuron pozicionirao podatak na sferi
prikaz_qnn = plot_bloch_multivector(stanje_nakon_ucenja)

plt.show()
prikaz_qnn
```

Šta predstavlja položaj ove strelice? 

Strelica se više ne nalazi na severnom polu, niti na ekvatoru. Ona je pod uticajem parametra nagnuta naniže za tačno 120 stepeni (odnosno 2.09 radijana) duž vertikalne y-ose.

Ko je odgovoran za njeno kretanje? 

Funkcija "kolo_qnn.ry(tezina_neurona, 0)" je simulirala rad kvantnog neurona, gde promenljiva tezina_neurona predstavlja ugao rotacije koji sistem fino podešava tokom procesa mašinskog učenja.

Kako sistem na osnovu ovoga donosi odluku? 

Kvantni klasifikator koristi hemisfere Blohove sfere kao logičke klase, gde severni pol predstavlja bezbedan paket (|0>), a južni pol mrežni napad (|1>). Pošto je strelica prešla ekvator i ušla duboko u južnu hemisferu, to je direktan dokaz da je naš kvantni AI uspešno identifikovao i klasifikovao podatak kao zlonameran upad u sistem.
