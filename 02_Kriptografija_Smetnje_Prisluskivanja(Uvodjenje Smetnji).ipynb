```python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

print("--- Simulacija anomalije i faznog šuma na Blohovoj sferi ---")

# 1. Pravimo novo kolo sa 1 kubitom
kolo_smetnja = QuantumCircuit(1)

# 2. Alisa šalje čist podatak (Hadamardova kapija stavlja kubit na x-osu)
kolo_smetnja.h(0)

# 3. TRENUTAK NAPADA: Alisa (ili Eva) prisluškuje kanal i uvodi anomaliju
# Rotiramo kubit oko Z-ose za uglove od pi/4 (45 stepeni) usled krađe informacija
ugao_smetnje = np.pi / 4
kolo_smetnja.rz(ugao_smetnje, 0)

# 4. Izvlačimo modifikovano kvantno stanje
stanje_smetnja = Statevector.from_instruction(kolo_smetnja)

# 5. Crta se nova Blohova sfera sa defektom
prikaz_smetnje = plot_bloch_multivector(stanje_smetnja)

plt.show()
prikaz_smetnje
```

Šta predstavlja skretanje strelice?

 Strelica više ne cilja direktno u tačku X. 
 Pomerila se za ugao od 45 stepeni, pi/4 radijana oko vertikalne z-ose.
 
 Ko je odgovoran za promenu? 
 
 Funkcija "kolo_smetnja.rz()" je simulirala energetski ili informacioni uticaj napadača (Eve) na kvantni komunikacioni kanal.
 
 Kako Bob zna da ga prisluškuju? Kada Bob na prijemnoj stanici u izmeri stanje ovog kubita, on dobija promenu u kvantnoj fazi.
 
 Pošto zna da je Alisa trebalo da pošalje čist podatak na x-osi, ovaj otklon mu jasno govori da je na liniji prisluškivač. Sistem automatski proglašava ključ kompromitovanim i prekida slanje.
