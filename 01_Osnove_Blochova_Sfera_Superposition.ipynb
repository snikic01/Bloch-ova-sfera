```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

print("--- Kreiranje stanja na Blohovoj sferi ---")

# 1. Pravimo kvantno kolo sa 1 kubitom
# Po defaultu, kubit pocinje na Severnom polu, tj. u stanju |0>
kolo = QuantumCircuit(1)

# 2. Primenjujemo Hadamardovu kapiju (H)
# Ova kapija uzima strelicu sa Severnog pola i rotira je za 90 stepeni pravo na Ekvator (X-osu)
kolo.h(0)

# 3. Izvlačimo matematički "Statevector" (vektor stanja) iz našeg kola
stanje = Statevector.from_instruction(kolo)

# 4. CRTANJE BLOHOV-e SFERE
# Ova funkcija uzima matematiku i pretvara je u 3D grafički prikaz
prikaz = plot_bloch_multivector(stanje)

# Prikazujemo grafiku na ekranu sveske
plt.show()
prikaz # Ova linija osigurava da Jupyter prikaže 3D loptu
```

Gde je startna pozicija? 

Kada se kvantno kolo pokrene, strelica se nalazi na samom vrhu lopte, na severnom polu, u stanju |0>.

Šta je uradila linija kolo.h(0)? 

Hadamardova kapija je uzela tu strelicu i zarotirala je za 90 stepeni naniže. 

Strelica je pala tačno na ekvator i sada cilja u tačku x.

Matematičko značenje stanja sa ekrana: 
U ovoj tački, verovatnoća da je kubit nula je tačno 50%, a da je jedinica je takođe 50%. Podatak je u stanju čiste, stabilne superpozicije.
